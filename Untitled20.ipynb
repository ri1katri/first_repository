```python
def check_input(element):
    return isinstance(element, str)

def get_substring(s):
    longest = ''
    current = ''

    for char in s:
        if char in current:
            current = current[current.index(char) +1:]
        current += char

        if len(current) > len(longest):
            longest = current
    return longest

def main(a):
    max_len = -1
    max_index = -1
    longest_string =''

    for i in range(len(a)):
        if check_input(a[i]):
            if len(a[i]) > max_len:
                max_len = len(a[i])
                max_index = i
                longest_string = a[i]

    if max_index == -1:
       return(-1,'')

    else:
        substring = get_substring(longest_string)
        return (max_index, substring)
```
